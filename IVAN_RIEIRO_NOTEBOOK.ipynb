```python
import numpy as np
import matplotlib.pyplot as graf
import pandas as pd
```

```python
TeslaStocks = pd.read_csv('TESLA.csv')

TeslaStocks = TeslaStocks.iloc[2:]

TeslaStocks.reset_index(drop=True, inplace=True)

TeslaStocks
```

```python

def formatear_valor(valor):
    if isinstance(valor, float):
        return round(valor, 2)
    return valor


columnas_numericas = TeslaStocks.columns.difference(['Date'])
TeslaStocks[columnas_numericas] = TeslaStocks[columnas_numericas].applymap(formatear_valor)
TeslaStocks
```

```python
TeslaStocks['Date'] = pd.to_datetime(TeslaStocks['Date'])

# Crear un diccionario de DataFrames agrupados por mes
dataframes_por_mes = {}
for month, data in TeslaStocks.groupby(TeslaStocks['Date'].dt.month):
    dataframes_por_mes[month] = data

for month in range(10, 13):
    print(f"Mes {month}/2021 \n{dataframes_por_mes[month]}")

for month in range(1, 10):
    print(f"Mes {month}/2022 \n{dataframes_por_mes[month]}")
```

```python
# Crear diccionarios vacÃ­os para almacenar los promedios
PromOpen = {}
PromClose = {}
PromHigh = {}
PromLow = {}
PromVol = {}

# Calcular los promedios para cada mes
for month in range(1, 13):
    PromOpen[month] = round(dataframes_por_mes[month]['Open'].mean(), 2)
    PromClose[month] = round(dataframes_por_mes[month]['Close'].mean(), 2)
    PromHigh[month] = round(dataframes_por_mes[month]['High'].mean(), 2)
    PromLow[month] = round(dataframes_por_mes[month]['Low'].mean(), 2)
    PromVol[month] = round(dataframes_por_mes[month]['Volume'].mean(), 2)

for month in range(1, 13):
    print(f"Mes {month}:")
    print(f"  Promedio de 'Open' = {PromOpen[month]}")
    print(f"  Promedio de 'Close' = {PromClose[month]}")
    print(f"  Promedio de 'High' = {PromHigh[month]}")
    print(f"  Promedio de 'Low' = {PromLow[month]}")
    print(f"  Promedio de 'Volume' = {PromVol[month]}")
    print()
```

```python
# Datos de apertura de los meses y nombres de los meses
Opens = [PromOpen[10], PromOpen[11], PromOpen[12], PromOpen[1], PromOpen[2], PromOpen[3], PromOpen[4], PromOpen[5], PromOpen[6], PromOpen[7], PromOpen[8], PromOpen[9]]
OpensMes = ["Oct 2021", "Nov 2021", "Dic 2021", "Ene 2022", "Feb 2022", "Mar 2022", "Abr 2022", "May 2022", "Jun 2022", "Jul 2022", "Ago 2022", "Sept 2022"]

# Crear un objeto de figura y ejes
fig, ax = graf.subplots()

# Crear el grÃ¡fico de barras
ax.bar(OpensMes, Opens)

# Rotar las etiquetas del eje x para una mejor legibilidad
graf.xticks(rotation=45, ha="right")

# Configurar las etiquetas y tÃ­tulo
ax.set_xlabel('Mes')
ax.set_ylabel('Apertura Promedio')
ax.set_title('Apertura Promedio de los Meses de acciones de Tesla')

# Mostrar el grÃ¡fico
graf.tight_layout()
graf.show()
```
